# Define the model

```python
#vectorize the input 
net = tflearn.input_data(shape=[None, 784])
#hidden layer_1
net = tflearn.fully_connected(net, 100, activation = 'relu')
#hidden layer_2
net = tflearn.fully_connected(net, 100,activation = 'relu')
#hidden layer_3
net = tflearn.fully_connected(net, 50, activation = 'relu')
#output layer
net = tflearn.fully_connected(net, 10, activation = 'softmax')

#lossfunction / update layer
net = tflearn.regression(net, loss='categorical_crossentropy', optimizer = 'adam')
#define model

model = tflearn.DNN(net)
```

# Train the model

```python
tf.reset_default_graph()
model.fit(x, y, n_epoch=3,batch_size=10, show_metric=True)
```

```python
test_x=data_test._images
test_y=data_test._labels
pre=model.predict(test_x)
for (x,y) in zip(pre,test_y):
    print("label = {} ,predicted = {}".format(np.argmax(y),np.argmax(x)))
```
